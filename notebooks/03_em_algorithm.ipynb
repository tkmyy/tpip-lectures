## ①
遺伝子頻度 $\boldsymbol{\theta}=\lbrack \theta_A, \theta_B, \theta_O \rbrack$ とした時の1人分の同時分布を以下に示す.

|遺伝型 \ 表現型|A|B|AB|O|
|:--:|:--:|:--:|:--:|:--:|
|**AA**|$$\theta_A^2$$|0|0|0|
|**AO**|$$2\theta_A\theta_O$$|0|0|0|
|**BB**|0|$$\theta_B^2$$|0|0|
|**BO**|0|$$2\theta_B\theta_O$$|0|0|
|**AB**|0|0|$$2\theta_A\theta_B$$|0|
|**OO**|0|0|0|$$\theta_O^2$$|

## ②
$h_i, g_i$ をそれぞれ $i$ 番目の表現型, 遺伝型とすると, 全員分の同時分布 $p\left(\left\{h_i, g_i \right\}_{i=1}^{N}; \boldsymbol{\theta}\right)$ は次のように表せる.

$$
\begin{align}
p\left(\left\{h_i, g_i \right\}_{i=1}^{N}; \boldsymbol{\theta}\right) &= p(\boldsymbol{\theta})
\prod_{i=1}^N p\left(h_i, g_i \mid \boldsymbol{\theta}\right) \\
\end{align}
$$

## ③
未知パラメータ $\boldsymbol{\theta}$ の第2種最尤推定

$$
\hat{\boldsymbol{\theta}}:=\underset{\boldsymbol{\theta}}{\operatorname{argmax}} p(\mathbf{h}; \boldsymbol{\theta})
$$

は, 適当な初期値 $\boldsymbol{\theta}^{(0)}$ を設定して次の漸化式の収束値として求めることができる.

$$
\boldsymbol{\theta}^{(t+1)}:=\underset{\boldsymbol{\theta}}{\operatorname{argmin}}\langle-\ln p(\mathbf{h}, \mathbf{g} ; \boldsymbol{\theta})\rangle_{p\left(\mathbf{g} \mid \mathbf{h} ; \boldsymbol{\theta}^{(t)}\right)}
$$

argmin の中身について期待値の線形性より,

$$
\begin{align}
\langle-\ln p(\mathbf{h}, \mathbf{g} ; \boldsymbol{\theta})\rangle_{p\left(\mathbf{g} \mid \mathbf{h} ; \boldsymbol{\theta}^{(t)}\right)} &= \left\langle -\ln p(\boldsymbol{\theta}) \prod_{i=1}^N p\left(h_i, g_i \mid \boldsymbol{\theta}\right) \right\rangle_{p\left(\mathbf{g} \mid \mathbf{h} ; \boldsymbol{\theta}^{(t)}\right)} \\
&= \left\langle -\ln p(\boldsymbol{\theta}) - \sum_{i=1}^N \ln p\left(h_i, g_i \mid \boldsymbol{\theta}\right) \right\rangle_{p\left(\mathbf{g} \mid \mathbf{h} ; \boldsymbol{\theta}^{(t)}\right)} \\
&= -\ln p(\boldsymbol{\theta}) - \sum_{i=1}^N \left\langle \ln p\left(h_i, g_i \mid \boldsymbol{\theta}\right) \right\rangle_{p\left(g_i \mid h_i ; \boldsymbol{\theta}^{(t)}\right)}
\end{align}
$$

第2項について①の表を元に計算すると,

$$
\begin{align}
\begin{split}
\sum_{i=1}^N \left\langle \ln p\left(h_i, g_i \mid \boldsymbol{\theta}\right) \right\rangle_{p\left(g_i \mid h_i ; \boldsymbol{\theta}^{(t)}\right)} &= \sum_{i=1}^N \sum_{g_i} p\left(g_i \mid h_i ; \boldsymbol{\theta}^{(t)}\right) \ln p\left(h_i, g_i \mid \boldsymbol{\theta}\right) \\
&= \sum_{i=1}^N \left\lbrace R_A\frac{\theta_A^{(t)^2}}{\theta_A^2 + 2\theta_A^{(t)}\theta_O^{(t)}} \ln \theta_A^2 + R_A\frac{2\theta_A^{(t)}\theta_O^{(t)}}{\theta_A^{(t)^2} + 2\theta_A^{(t)}\theta_O^{(t)}} \ln \left(2\theta_A\theta_O \right) \right. \\
&\left. \quad\quad\quad + R_B\frac{\theta_B^{(t)^2}}{\theta_B^2 + 2\theta_B^{(t)}\theta_O^{(t)}} \ln \theta_B^2 + R_B\frac{2\theta_B^{(t)}\theta_O^{(t)}}{\theta_B^{(t)^2} + 2\theta_B^{(t)}\theta_O^{(t)}} \ln \left(2\theta_B\theta_O \right) \right. \\
&\left. \quad\quad\quad + R_{AB}\cdot1\cdot\ln \left (2 \theta_A\theta_B\right) \right. \\
&\left. \quad\quad\quad + R_O\cdot1\cdot\ln \theta_O^2 \right\rbrace \\
&= N \left\lbrace R_A\frac{2\theta_A^{(t)^2}}{\theta_A^{(t)^2} + 2\theta_A^{(t)}\theta_O^{(t)}} \ln \theta_A + R_A\frac{2\theta_A^{(t)}\theta_O^{(t)}}{\theta_A^{(t)^2} + 2\theta_A^{(t)}\theta_O^{(t)}} \left(\ln2 + \ln\theta_A+\ln\theta_O \right) \right. \\
&\left. \quad\quad\quad + R_B\frac{2\theta_B^{(t)^2}}{\theta_B^2 + 2\theta_B^{(t)}\theta_O^{(t)}} \ln \theta_B + R_B\frac{2\theta_B^{(t)}\theta_O^{(t)}}{\theta_B^{(t)^2} + 2\theta_B^{(t)}\theta_O^{(t)}} \left(\ln2 + \ln\theta_B+\ln\theta_O \right) \right. \\
&\left. \quad\quad\quad + R_{AB}\left(\ln2 + \ln\theta_A+\ln\theta_B \right) + 2R_O\ln\theta_O  \right\rbrace \\
&= N \left\lbrace \left(R_A\frac{2\theta_A^{(t)^2}+2\theta_A^{(t)}\theta_O^{(t)}}{\theta_A^{(t)^2} + 2\theta_A^{(t)}\theta_O^{(t)}} + R_{AB} \right)\ln\theta_A
+ \left(R_B\frac{2\theta_B^{(t)^2}+2\theta_B^{(t)}\theta_O^{(t)}}{\theta_B^{(t)^2} + 2\theta_B^{(t)}\theta_O^{(t)}} + R_{AB} \right)\ln\theta_B \right. \\
&\left. \quad\quad\quad + \left(R_A\frac{2\theta_A^{(t)}\theta_O^{(t)}}{\theta_A^{(t)^2} + 2\theta_A^{(t)}\theta_O^{(t)}} + R_B\frac{2\theta_B^{(t)}\theta_O^{(t)}}{\theta_B^{(t)^2} + 2\theta_B^{(t)}\theta_O^{(t)}} + 2R_O \right)\ln\theta_O + const \right\rbrace \\
\end{split}
\end{align}
$$

ここで, $\ln\theta_A, \ln\theta_B, \ln\theta_O$ の係数をそれぞれ $\alpha, \beta, \gamma$ とすると, 更新式は以下のように表せる.

$$
\begin{align}
\boldsymbol{\theta}^{(t+1)} &= \underset{\boldsymbol{\theta}}{\operatorname{argmin}}\left\lbrace -\ln p(\boldsymbol{\theta}) -N(\alpha\ln\theta_A+\beta\ln\theta_B+\gamma\ln\theta_O+const)\right\rbrace \\
&= \underset{\boldsymbol{\theta}}{\operatorname{argmin}}\left\lbrace -N(\alpha\ln\theta_A+\beta\ln\theta_B+\gamma\ln\theta_O)+const\right\rbrace \\
&= \underset{\boldsymbol{\theta}}{\operatorname{argmin}}\left\lbrack -N\left\lbrace\alpha\ln\theta_A+\beta\ln\theta_B+\gamma\ln{(1-\theta_A-\theta_B)}\right\rbrace+const\right\rbrack \quad\because \theta_A+\theta_B+\theta_O=1 
\end{align}
$$

$f(\boldsymbol\theta)=-N\left\lbrace\alpha\ln\theta_A+\beta\ln\theta_B+\gamma\ln{(1-\theta_A-\theta_B)}\right\rbrace$ とおくと, $\theta_A, \theta_B$ についてそれぞれ微分して勾配が0となる点が解の候補となるので,

$\theta_A$ について,

$$
\begin{align}
\frac{\partial f(\boldsymbol\theta)}{\partial \theta_A} = -N \left(\alpha \frac{1}{\theta_A} + \gamma \frac{-1}{1-\theta_A-\theta_B}\right) &= 0 \\
\therefore \frac{\alpha}{\theta_A} &= \frac{\gamma}{1-\theta_A-\theta_B} \\
(\alpha+\gamma)\theta_A + \alpha\theta_B &= \alpha
\end{align}
$$

$\theta_B$ についても同様に,

$$
\begin{align}
\frac{\partial f(\boldsymbol\theta)}{\partial \theta_B} = -N \left(\beta \frac{1}{\theta_B} + \gamma \frac{-1}{1-\theta_A-\theta_B}\right) &= 0 \\
\therefore \frac{\beta}{\theta_B} &= \frac{\gamma}{1-\theta_A-\theta_B} \\
\beta\theta_A + (\beta+\gamma)\theta_B &= \beta
\end{align}
$$

2式より,

$$
\begin{align}
(\alpha+\beta+\gamma)\theta_A + (\alpha+\beta+\gamma)\theta_B &= \alpha + \beta \\ 
\theta_A + \theta_B  = \frac{\alpha + \beta}{\alpha + \beta + \gamma}
\end{align}
$$

$$
\begin{align}
(\alpha+\gamma)\theta_A + \left(\frac{\alpha + \beta}{\alpha + \beta + \gamma} - \theta_A \right) &= \alpha \\
\gamma \theta_A &= \alpha-\frac{\alpha(\alpha+\beta)}{\alpha + \beta + \gamma}
\end{align}
$$

整理すると, 更新後のパラメータ $\boldsymbol\theta^{(t+1)}$ は $\alpha, \beta, \gamma$ を用いて次のように求まる.

$$
\begin{align}
\theta_A &= \frac{\alpha}{\alpha+\beta+\gamma} \\
\theta_B &= \frac{\beta}{\alpha+\beta+\gamma} \\
\theta_O &= \frac{\gamma}{\alpha+\beta+\gamma}
\end{align}
$$

$$
\begin{align}
\alpha &= 2R_A\frac{\theta_A^{(t)}+\theta_O^{(t)}}{\theta_A^{(t)} + 2\theta_O^{(t)}} + R_{AB} \\
\beta &= 2R_B\frac{\theta_B^{(t)}+\theta_O^{(t)}}{\theta_B^{(t)} + 2\theta_O^{(t)}} + R_{AB} \\
\gamma &= 2R_A\frac{\theta_O^{(t)}}{\theta_A^{(t)} + 2\theta_O^{(t)}} + 2R_B\frac{\theta_O^{(t)}}{\theta_B^{(t)} + 2\theta_O^{(t)}} + 2R_O
\end{align}
$$

## ④
③で導出した更新式を用いて $\boldsymbol{\theta}$ の最尤解を求める. 初期値は以下の5通りとし, それぞれ20回更新を行なった.

||$$\theta_A^{(0)}$$|$$\theta_B^{(0)}$$|$$\theta_O^{(0)}$$|
|--|--|--|--|
|**1**|0.1|0.2|0.7|
|**2**|0.2|0.5|0.3|
|**3**|0.3|0.3|0.4|
|**4**|0.5|0.2|0.3|
|**5**|0.2|0.1|0.7|

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 16


def run_em(theta_A, theta_B, theta_O, n_iter=20):
    print("---初期値---")
    print(f"theta_A: {round(theta_A, 3)}")
    print(f"theta_B: {round(theta_B, 3)}")
    print(f"theta_O: {round(theta_O, 3)}")

    R_dic = {"A": 0.4, "B": 0.2, "AB": 0.1, "O": 0.3}
    res = [
        {
            "iteration": 0,
            "theta_A": theta_A,
            "theta_B": theta_B,
            "theta_O": theta_O
        }
    ]
    for i in range(n_iter):
        alpha = R_dic["AB"] + 2*(R_dic["A"]*(theta_A+theta_O))/(theta_A+2*theta_O)
        beta = R_dic["AB"] + 2*(R_dic["B"]*(theta_B+theta_O))/(theta_B+2*theta_O)
        gamma = 2*R_dic["O"] + 2*theta_O*((R_dic["A"]/(theta_A+2*theta_O))+(R_dic["B"]/(theta_B+2*theta_O)))

        theta_A = alpha/(alpha+beta+gamma)
        theta_B = beta/(alpha+beta+gamma)
        theta_O = gamma/(alpha+beta+gamma)

        res += [
            {
                "iteration": i+1,
                "theta_A": theta_A,
                "theta_B": theta_B,
                "theta_O": theta_O
            }
        ]

    print("---最尤解---")
    print(f"theta_A: {round(theta_A, 3)}")
    print(f"theta_B: {round(theta_B, 3)}")
    print(f"theta_O: {round(theta_O, 3)}")

    return pd.DataFrame(res)


def show_results(df):
    plt.figure(figsize=(10, 3))
    for col in ("theta_A", "theta_B", "theta_O"):
        plt.plot(df["iteration"].astype(str), df[col], marker="*", markersize=8, label=col)

    plt.xlabel("更新回数")
    plt.ylim(0, 0.8)
    plt.legend(bbox_to_anchor=(1.02, 1.02), loc="upper left")
    plt.grid()
    plt.show()


# パラメータの初期値
theta_l = [
    [0.1, 0.2],
    [0.2, 0.5],
    [0.3, 0.3],
    [0.5, 0.2],
    [0.2, 0.1]
]
for theta_A, theta_B in theta_l:
    theta_O = 1 - (theta_A + theta_B)
    res_df = run_em(theta_A, theta_B, theta_O)
    show_results(res_df)
```

以上より, 初期値による収束値のばらつきがないため, 遺伝子頻度 $\boldsymbol{\theta}$ の最尤解は

$$
\begin{align}
\hat{\theta_A} &= 0.292\\
\hat{\theta_B} &= 0.163\\
\hat{\theta_O} &= 0.545\\
\end{align}
$$

と求まった.
